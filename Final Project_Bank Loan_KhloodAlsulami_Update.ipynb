```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

2- <font color='#6a5acd'><b>Train Test Splitting</b></font>

```python
# Training Set and Test Set Preparation and Splitting 
X = df.drop(['Personal Loan', 'ZIP Code', 'Mortgage', 'Online', 'ID'], axis=1)
y = df['Personal Loan']
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100)
```

Note: To perform logistic regression, the independent variables should primarily be numerical factors, such as age, education level, years of experience, family size, and income. These variables should be in numerical form to ensure accurate analysis. However, categorical variables like credit card should be encoded into numerical values before applying the logistic regression model.

3- <font color='#6a5acd'><b>Convert Categorical Data to Numeric</b></font>
    ** Convert CCAvg to numeric 

```python
from sklearn.preprocessing import LabelEncoder

# Assuming X_train is your DataFrame
non_numeric_columns = X_train.select_dtypes(include=['object', 'category']).columns

le = LabelEncoder()
for col in non_numeric_columns:
    X_train[col] = le.fit_transform(X_train[col])
```

```python
def safe_convert_to_float(value):
    if isinstance(value, str) and '/' in value:
        try:
            numerator, denominator = map(float, value.split('/'))
            return numerator / denominator if denominator != 0 else 0
        except ValueError:
            return None  # Handle cases where conversion fails
    return value
X_test['CCAvg'] = X_test['CCAvg'].apply(safe_convert_to_float)
```

4- <font color='#6a5acd'><b>Run the Model</b></font>

```python
# Fit Model with Training Set
lr_model = LogisticRegression(max_iter=2000, penalty='l2', C=1.0, solver='liblinear')
# Fitting the model to the training data
lr_model.fit(X_train, y_train)
```

```python
# Predicting the values for the testing data
y_pred = lr_model.predict(X_test)
```

5- <font color='#6a5acd'><b>Check the Model Accuracy</b></font>

```python
from sklearn.metrics import accuracy_score
# Evaluate the Model's Performance Using Metric
print('Accuracy:', accuracy_score(y_test, y_pred))
```

7- <font color='#6a5acd'><b> Evaluate the model's performance using alternative metrics, such as Precision, Recall, and F1-score</b></font>

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate Recall
recall_micro = recall_score(y_test, y_pred, average='micro')
recall_macro = recall_score(y_test, y_pred, average='macro')
recall_weighted = recall_score(y_test, y_pred, average='weighted')

# Calculate Precision
precision_micro = precision_score(y_test, y_pred, average='micro')
precision_macro = precision_score(y_test, y_pred, average='macro')
precision_weighted = precision_score(y_test, y_pred, average='weighted')

# Calculate F1-Score
f1score_micro = f1_score(y_test, y_pred, average='micro')
f1score_macro = f1_score(y_test, y_pred, average='macro')
f1score_weighted = f1_score(y_test, y_pred, average='weighted')

# Print the Results
print(f'Recall (micro): {recall_micro}')
print(f'Recall (macro): {recall_macro}')
print(f'Recall (weighted): {recall_weighted}')

print(f'Precision (micro): {precision_micro}')
print(f'Precision (macro): {precision_macro}')
print(f'Precision (weighted): {precision_weighted}')

print(f'F1-Score (micro): {f1score_micro}')
print(f'F1-Score (macro): {f1score_macro}')
print(f'F1-Score (weighted): {f1score_weighted}')
```

To convert 0/30 to float
